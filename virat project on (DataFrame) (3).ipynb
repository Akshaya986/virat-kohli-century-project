```python
import matplotlib.pyplot as plt
```

```python
x=df["country"].head()
y=df["Strike Rate"].head().sort_values(ascending=False)

plt.bar(x,y)
plt.xlabel("country")
plt.ylabel("Strike Rate")
plt.title("country wise strike rate")
plt.show()
```

```python
x=df["country"].head()
y=df["Strike Rate"].head().sort_values(ascending=False)

plt.barh(x,y)
plt.xlabel("country")
plt.ylabel("Strike Rate")
plt.title("country wise strike rate")
plt.show()
```

```python
x=df["country"].head()
y=df["Strike Rate"].head()

plt.plot(x,y)
plt.xlabel("country")
plt.ylabel("Strike Rate")
plt.title("country wise strike rate")
plt.show()
```

```python
x=df["country"].head(3)
y=df["Strike Rate"].head(3)

plt.figure(figsize=(6,8))
plt.pie(y,labels=x,startangle=0,explode=[0.1,0,0,],shadow=False,colors=["green","yellow","blue"])
plt.legend(title="country wise strike rate")
plt.show()
```
